```python
import cv2
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy.io
from mpl_toolkits.mplot3d import Axes3D
import numpy.linalg
import random

%matplotlib inline
```

```python
def to_hom_coords(coords):
    return np.hstack((coords, 1))

def from_hom_coords(coords):
    return coords[:-1] / (1.0 * coords[-1]);
```

### 1) Оценка фундаментальной матрицы.

```python
matfile =  scipy.io.loadmat('estimateF.mat')
print matfile.keys(), '\n', matfile['x'].shape
first_projection = matfile['x'][0, 0]
second_projection = matfile['x'][1, 0]
print first_projection.shape, second_projection.shape
```

Считали данные для точек с изображений, визуализируем эти точки.

```python
def show_projection(image, projection):
    delta = 7
    for i in xrange(projection.shape[1]):
        cv2.circle(image, (int(projection[0, i]), int(projection[1, i])), delta, (255, 0, 0), thickness = -1)
    plt.imshow(image)
    plt.show()
    
image_first = cv2.imread('im1.JPG')
image_second = cv2.imread('im2.JPG')

show_projection(image_first.copy(), first_projection)
show_projection(image_second.copy(), second_projection)
```

Найдем матрицы нормализации и применим их. Третью однородную координату не изменяем, ее дисперсия нулевая, так что нормализовать не получится и не надо.

```python
def find_norm_matrix(coords):
    matrix = np.zeros((3, 3))
    matrix[2, 2] = 1
    for i in xrange(2):
        coord_values = coords[i]
        mean_coord = np.mean(coord_values)
        shifted_values = coord_values[:] - mean_coord
        norm_value = numpy.linalg.norm(shifted_values)
        matrix[i, i] = 1 / norm_value
        matrix[i, 2] = - mean_coord / norm_value;
    return matrix

N_1 = find_norm_matrix(first_projection)
norm_first_projection = np.dot(N_1, first_projection)

N_2 = find_norm_matrix(second_projection)
norm_second_projection = np.dot(N_2, second_projection)

print N_1, '\n', np.mean(norm_first_projection[1]), numpy.linalg.norm(norm_first_projection[1]), '\n'
print N_2, '\n', np.mean(norm_second_projection[0]), numpy.linalg.norm(norm_second_projection[0])
```

Теперь необходимо оценить матрицу F по нормализованным проекциям.

```python
def find_F(projection_1, projection_2):    
    A = np.zeros((0, 8))
    b = np.ones(projection_1.shape[1])
    b = -1 * b
    for i in xrange(projection_1.shape[1]):
        first_u = projection_1[:, i]
        second_v = projection_2[:, i]
        coefs = np.zeros((1, 8))
        coefs[0, 0] = first_u[0] * second_v[0]
        coefs[0, 1] = first_u[0] * second_v[1]
        coefs[0, 2] = first_u[0]        
        coefs[0, 3] = first_u[1] * second_v[0]
        coefs[0, 4] = first_u[1] * second_v[1]
        coefs[0, 5] = first_u[1]
        coefs[0, 6] = second_v[0]
        coefs[0, 7] = second_v[1]          
        A = np.vstack((A, coefs))
    print A.shape, b.shape, '\n'
    res = numpy.linalg.lstsq(A, b)
    result = numpy.append(res[0].copy(), 1.0)
    return result.reshape((3, 3))


F_in = find_F(norm_first_projection, norm_second_projection)
print F_in, '\n'

F_trans = np.dot(np.dot(N_1.transpose(), F_in), N_2)
print F_trans
```

Приводим матрицу F к рангу 2.

```python
u, s, v = numpy.linalg.svd(F_trans)
min_sing_val = min(s)
print s, '\n'
for i in xrange(len(s)):
    if s[i] == min_sing_val:
        s[i] = 0
        
F =  np.dot(np.dot(u, np.diag(s)), v)
F = F / (1.0 * F[2,2])
print F, '\n', "F matrix rank -",  np.linalg.matrix_rank(F)
```

Найдем эпиполярные линии, соответствующие всем заданным точкам первого изображения.

```python
def find_epipolar_lines(projection, F):
    lines = projection.copy()
    for i in xrange(projection.shape[1]):
        point = projection[:, i]
        lines[:, i] = np.dot(F.transpose(), point)
    return lines
     
epipolar_lines = find_epipolar_lines(first_projection, F)
print epipolar_lines.shape, epipolar_lines[:, 0]
```

Визуализируем 20 случайных эпиполярных линий и соответствующие им вторые точки из заданных пар на втором изображении.

```python
matplotlib.rc("savefig", dpi=140)
def show_epipoles(image, projection, lines):
    delta = 12
    for r in xrange(20):
        i = int(random.random() * projection.shape[1])
        cv2.circle(image, (int(projection[0, i]), int(projection[1, i])), delta, (255, 0, 0), thickness = -1)       
        line = lines[:, i]
        a = line[0]
        b = line[1]
        c = line[2]        
        cv2.line(image, (0, int(-c/b)), (image.shape[1], int((-c - image.shape[1] * a)/b)), (0, 255, 0), thickness = 6)
    plt.imshow(image)
    plt.show()

show_epipoles(image_second.copy(), second_projection, epipolar_lines)
```

Эпиполярные линии лежат достаточно близко к точкам, ура.

### 2) Оценка камеры матрицы

Оцениваем значение b.

```python
b = numpy.linalg.solve(F.transpose()[:2, :2], -F.transpose()[:2, 2])
b = np.append(b, 1.0)
print b#, '\n', F, '\n',  F.transpose()[:2, 2], '\n', F.transpose()[:2, :2]
```

Из оцененого b получаем A.

```python
def make_cross_product_matrix(vector):
    matrix = np.zeros((3, 3))
    matrix[0, 1] = -vector[2]
    matrix[0, 2] = vector[1]
    matrix[1, 2] = -vector[0]
    matrix[1, 0] = vector[2]
    matrix[2, 0] = -vector[1]
    matrix[2, 1] = vector[0]
    return matrix

A = -np.dot(make_cross_product_matrix(b), F)
print A#, '\n', make_cross_product_matrix(b)
```

Составляем данные, выводим матрицы обеих камер.

```python
P_1 = np.zeros((3, 4))
for i in xrange(3):
    P_1[i, i] = 1
print "Матрица первой камеры \n", P_1, '\n'
P_2 = np.hstack((A, b.reshape((3, 1))))
print "Матрица второй камеры \n", P_2
```

### 3) Триангуляция

Ищем 3-D координаты, соответствующие каждой паре точек.

```python
X_list = np.zeros((4, 0))
P1_T = P_1.transpose()
P2_T = P_2.transpose()
#print P1_T, '\n', P2_T
for i in xrange(first_projection.shape[1]):
    first = first_projection[:, i]
    second = second_projection[:, i]
    A = np.zeros((4, 4))
    b = np.zeros((4, 1))
   # X[3] = 1 ограничение зададим с помощью ненулевой b, выражение для А из лекции
    A[:, 0] = first[0] * P1_T[:, 2] - P1_T[:, 0]
    A[:, 1] = first[1] * P1_T[:, 2] - P1_T[:, 1]
    A[:, 2] = second[0] * P2_T[:, 2] - P2_T[:, 0]
    A[:, 3] = second[1] * P2_T[:, 2] - P2_T[:, 1]
    b = - A[3, :]  
#     print A[:3, :].transpose() ,'\n',  b
#     break
    res = numpy.linalg.lstsq(A[:3, :].transpose(), b)[0]
    res = np.append(res, 1.0)
    res_norm = numpy.linalg.norm(res)
    res /= res_norm
    #print res, numpy.linalg.norm(res), res.shape
    X_list = np.hstack((X_list, res.reshape((4, 1))))
```

Попробуем визуализировать полученные точки
