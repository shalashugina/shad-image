```python
import cv2
from skimage import io
import numpy
from matplotlib import pyplot
%matplotlib inline
```

#### 1. Считываем изображение, создаем черно-белый вариант, смотрим на гистограмму.

```python
image = skimage.io.imread("room.jpg")
grey_image = skimage.color.rgb2grey(image).astype('float32')
print "Значения яркости -",skimage.dtype_limits(grey_image)

hist = cv2.calcHist([grey_image],[0],None,[256],[0.,1.])

pyplot.plot(hist)
pyplot.xlim([0,256])
print "Гистограмма:"
pyplot.show()
```

Диапазон яркостей довольно узкий (темное изображение), значит для повышения контрастности должна хорошо работать линейная коррекция.

#### 2. Линейная коррекция изображения.

```python
image_grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
lineary_corrected_image = cv2.equalizeHist(image_grey)
hist_initial = cv2.calcHist([image_grey],[0],None,[256],[0,256])
hist_corrected = cv2.calcHist([lineary_corrected_image],[0],None,[256],[0,256])

skimage.io.imshow(lineary_corrected_image)
print "Скорректированное изображение"
skimage.io.show()

skimage.io.imshow(image_grey)
print "Исходное изображение"
skimage.io.show()

pyplot.plot(hist_corrected)
pyplot.xlim([0,256])
print "Гистограмма скорректированного изображения"
pyplot.show()

pyplot.plot(hist_initial)
pyplot.xlim([0,256])
print "Гистограмма исходного изображения"
pyplot.show()
```

Контрастность изображения повысилась.

#### 4. Получаем два изображения, зашумленные гауссовским шумом и шумом "соль и перец".

```python
gaussian_noise = image_grey.copy().astype('float32')
gaussian_noise += numpy.random.normal(0, 20, image_grey.shape)
    
skimage.io.imshow(gaussian_noise)
print "Изображение с гауссовым шумом"
skimage.io.show()
    
sap_noise = image_grey.copy()
for x in range(0, sap_noise.shape[0]):
    for y in range(0, sap_noise.shape[1]):
        noise = numpy.random.random()
        if noise < 0.03:
            sap_noise[x][y] = 0
        elif noise > 0.97:
            sap_noise[x][y] = 255
    
skimage.io.imshow(sap_noise)
print "Изображение с шумом 'соль и перец'"
skimage.io.show()
```

#### 5. Применим гауссовскую фильтрацию к изображению, зашумленному гауссовским шумом.

```python
parameters = [(5, 1), (5, 2), (7, 1), (3, 1)]
for ksize, sigma in parameters:
    blurred_image = cv2.GaussianBlur(gaussian_noise, (ksize, ksize), sigma)
    skimage.io.imshow(blurred_image)
    print "После гауссовской фильтрации с параметрами размер фильтра и сигма равными", ksize, " и ", sigma
    skimage.io.show()
```

Для сравнительно небольшого шума для заметного сглаживания хватает параметра сигма равного единице, при сигма равном 2 начинает происходить излишнее сглаживание и размытие границ. Для сигма = 1 по эмпирическому правилу подходит размер ядра = 5, но по изображениям кажется, что второй параметр влияет не так существенно.
