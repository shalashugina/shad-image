```python
import os
import skimage.io
import skimage.feature
import numpy as np
import sklearn
from sklearn import svm
from sklearn import cross_validation
import cv2
%matplotlib inline
```

##  Детектирование лиц методом скользящего окна.

### Извлечение HOG-признаков.

```python
positive_filenames = os.listdir("train/faces/")
negative_filenames = os.listdir("train/non-faces/")
filenames = negative_filenames + positive_filenames
print "Число примеров лиц -", len(positive_filenames),"; число примеров фона -", len(negative_filenames)

(a, b) = (8, 8)
X_features = np.zeros((0, 9*24*32/(a*b)))
y_classes = []

for name in positive_filenames:
    image = skimage.io.imread("train/faces/" + name, as_grey=True)
    hos = skimage.feature.hog(image, orientations=9, pixels_per_cell=(a, b), cells_per_block=(1, 1))
    X_features = np.vstack((X_features, hos))
    y_classes += [1]
                  
for name in negative_filenames:
    image = skimage.io.imread("train/non-faces/" + name, as_grey=True)
    hos = skimage.feature.hog(image, orientations=9, pixels_per_cell=(a, b), cells_per_block=(1, 1))
    X_features = np.vstack((X_features, hos))
    y_classes += [0]
    
print "Размер обучающих изображений", skimage.io.imread("train/faces/2.png").shape
(window_x, window_y) = (32, 24)
print "Размер массива признаков", X_features.shape
```

### Обучение SVM

```python
SVM_classifier = sklearn.svm.SVC()
SVM_classifier.fit(X_features, y_classes)
```

### Детектирование лиц

#### Детектирование лиц на одной картинке, использовалось для подбора параметров для "сложных" изображений.  Здесь оставлена самая "сложная" картинка, лицо никак не удалось распознать подбором параметров ячейки. Даже коленки определились раньше лица!

```python
answers = []
name_number = 1
test_1 = cv2.imread('test/4.jpg')
test_image = cv2.cvtColor(test_1, cv2.COLOR_BGR2GRAY)
for x in xrange(0, test_image.shape[0] - window_x):
    for y in xrange(0, test_image.shape[1] - window_y):
        subimage = test_image[x : x + window_x, y : y + window_y]
        hos = skimage.feature.hog(subimage, orientations=9, pixels_per_cell=(a, b), cells_per_block=(1, 1))
        is_face = SVM_classifier.predict(hos)
        if is_face:
            #print "Found",x, y, name_number
            answers += [(name_number, x, y)]
            
print len(answers)

color_test_1 = cv2.cvtColor(test_image, cv2.COLOR_GRAY2RGB)
for name_number, x, y in answers:
    cv2.rectangle(color_test_1,(y,x),(y+window_y,x+window_x),(255, 0, 0))
skimage.io.imshow(color_test_1)
skimage.io.show()
```
