### 1) Проективные преобразования.

Функция для визуализации отрезков.

```python
def plot_grid(startpoints, endpoints, steps_number):
    x_all = np.array([])
    y_all = np.array([])    
    for i in xrange(startpoints.shape[1]):
        x = np.linspace(startpoints[0, i], endpoints[0, i], steps_number)
        y = np.linspace(startpoints[1, i], endpoints[1, i], steps_number)
        x_all = np.hstack((x_all, x))
        y_all = np.hstack((y_all, y))
    plt.figure(figsize=(5, 5))
    plt.plot(x_all, y_all, 'r,')
    plt.show()
```

Визуализируем регулярную сетку

```python
import lines
plot_grid(lines.startpoints, lines.endpoints, 1000)
```

Функции перевода в однородные координаты и из них.

```python
def to_hom_coords(coords):
    return np.hstack((coords, 1))

def from_hom_coords(coords):
    return coords[:-1] / (1.0 * coords[-1]);
```

Функция для применения проективного преобразования

```python
def apply_transform(startpoints, endpoints, H):
    dim = startpoints.shape[0]
    new_startpoints = np.zeros((dim, 0))
    new_endpoints = np.zeros((dim, 0))
    for i in xrange(startpoints.shape[1]):
        new_h_startpoint = np.dot(H, to_hom_coords(startpoints[:, i]))
        new_h_endpoint = np.dot(H, to_hom_coords(endpoints[:, i]))
        new_startpoints = np.hstack((new_startpoints, from_hom_coords(new_h_startpoint).reshape(dim, 1)))
        new_endpoints = np.hstack((new_endpoints, from_hom_coords(new_h_endpoint).reshape(dim, 1)))
    return new_startpoints, new_endpoints
```

Выполним преобразования с заданными матрицами и визуализуем результат.

```python
transforms = [lines.H1, lines.H2, lines.H3, lines.H4]

for H in transforms:
    startpoints_transformed, endpoints_transformed = apply_transform(lines.startpoints, lines.endpoints, H)
    plot_grid(startpoints_transformed, endpoints_transformed, 5000)
```

#### 2-2) Вычислим положение и направление камер

```python
def null(a, rtol= 1e-5):
    u, s, v = np.linalg.svd(a)
    rank = (s > rtol * s[0]).sum()
    return v[rank :].T.copy()
```

### 3) Выполним калибровку камеры

```python
matfile =  scipy.io.loadmat('calibration.mat')
print matfile.keys(), '\n', matfile['x'].shape
all_coords = matfile['model']
first_projection = matfile['x'][0, 0]
second_projection = matfile['x'][0, 1]
print all_coords.shape, first_projection.shape, second_projection.shape
```

Посмотрим, координаты и проекции каких точек нам даны.

```python
def show_projection(image, projection):
    delta = 10
    for i in xrange(projection.shape[1]):
        cv2.circle(image, (int(projection[0, i]), int(projection[1, i])), delta, (0, 255, 255), thickness = -1)
    plt.imshow(image)
    plt.show()
    
image_first = cv2.cvtColor(cv2.imread('calib1.JPG'), cv2.COLOR_BGR2RGB)
image_second = cv2.cvtColor(cv2.imread('calib2.JPG'), cv2.COLOR_BGR2RGB)

show_projection(image_first.copy(), first_projection)
show_projection(image_second.copy(), second_projection)
```

Матрицу для камеры будет оценивать, используя соответствие всех точек в пространстве точкам их проекций на изображения и метод наименьших квадратов. Получаем из каждой точки x = PX три линейных уравнения на параметры, все невошедшие в уравнение P_ij считаем вошедшими с множителем 0. Будем искать P как решение Ax= b, где каждая строка A - вектор коэффициентов уравнения, а элемент b - значение соответствующей этому уравнению координаты проекции точки.

```python
def find_P(all_coords, projection):    
    A = np.zeros((0, 12))
    b = np.zeros(0)
    for point in xrange(all_coords.shape[1]):
        point_X = to_hom_coords(all_coords[:, point])
        for x_coord in range(3):
            coefs = np.zeros((1, 12))
            b = np.hstack((b, projection[x_coord, point]))
            for X_coord in range(4):
                coefs[:, 4 * x_coord + X_coord] = point_X[X_coord]
            A = np.vstack((A, coefs))
    # P, sum error
    res = lstsq(A, b)
    error = sqrt(res[1] /all_coords.shape[1])
    return res[0].reshape((3, 4)), error

P_first, error_first = find_P(all_coords, first_projection)
P_second, error_second = find_P(all_coords, second_projection)    
print '\n', "Error for first camera -", error_first, '\n', "Error for second camera -", error_second
```

Получили искомые матрицы P и оценку ошибки.

```python
def rq_function(a):
    (m, n) = a.shape
    e = np.eye(m)
    p = e[:, range(m - 1, -1, -1)]
    (q0, r0) = np.linalg.qr(np.dot(np.dot(p, a[:, 0:m].transpose()), p))

    r = np.dot(np.dot(p, r0.transpose()), p)
    q = np.dot(np.dot(p, q0.transpose()), p)

    fix = np.diag(np.sign(np.diag(r)))

    r = np.dot(r, fix)
    q = np.dot(fix, q)

    if n > m:
        add_q = np.dot(np.linalg.inv(r), a[:, m:n])
        q = np.vstack((q, np.dot(np.linalg.inv(r), add_q.transpose()[0])))
    return(r, q)

r1, q1 = rq_function(P_first)
r2, q2 = rq_function(P_second)

print "First camera", '\n', P_first, '\n', '\n',r1, '\n', '\n' ,q1, '\n', '\n'
print "Second camera", '\n', P_second, '\n', '\n', r2, '\n', '\n', q2
```

Получили, что матрица калибровки верхнедиагональная, как и должно быть.

Проверим также, насколько правдоподобна получившаяся матрица P, использовав ее для нахождения положения точек на изображениях.

```python
def dots_projection(image, P, all_coords):
    delta = 10
    for i in xrange(all_coords.shape[1]):
        coords = from_hom_coords(np.dot(P, to_hom_coords(all_coords[:, i])))
        cv2.circle(image, (int(coords[0]), int(coords[1])), delta, (0, 255, 255), thickness = -1)
    plt.imshow(image)
    plt.show()
    
dots_projection(image_first.copy(), P_first, all_coords)
dots_projection(image_second.copy(), P_second, all_coords)
```

Видно, что точки сместились из их изначальных положений, но при этом их расположение в целом сохранено неплохо. Легко понять, какой исходной точке соответствует каждая новая.
