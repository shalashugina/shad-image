```python
import skimage.io
import numpy
import sklearn
from sklearn.cluster import KMeans
import cv2
%matplotlib inline
```

## Кластеризация попиксельных признаков

```python
def clustering(image):
    width = image.shape[0]
    height = image.shape[1]
    image_copy = numpy.reshape(image, (-1, 1))
    clustering = KMeans(n_clusters=2)
    new_image = clustering.fit_predict(image_copy)
    print "First cluster center intensity -", clustering.cluster_centers_[0], "; Second cluster center intensity -", clustering.cluster_centers_[1]
    clusterized_image = numpy.reshape(new_image, (width, height))
    return clusterized_image
```

#### Задали функцию кластеризации пикселей изображения на два кластера по их интенсивности, применим ее к данным изображениям.

```python
images_names = ["data/1.png", "data/2.png", "data/3.png", "data/4.png"] 

for name in images_names:
    image = skimage.io.imread(name, as_grey=True)
    print "Initial image"
    skimage.io.imshow(image)
    skimage.io.show()
    new_image = clustering(image)
    print "Image after clustering"
    skimage.io.imshow(new_image)
    skimage.io.show()
```

#### Довольно хорошо сегментируются изображения, у которых основные объекты заметно отличаются по интенсивности от фона и интенсивность относительно слабо меняется в области самих объектов (вертолет и льдина). Чуть хуже выделился дом, так как он контрастен на фоне неба, но по интенсивности близок к траве, дом и трава объединились в один кластер. Ракушка на изображении выделена частично, так как заметно неоднородна, некоторые ее участки заметно светлее других и сливаются с фоном(песком). В итоге такая простая кластеризация по эффекту похожа на бинаризацию (с условными значениями черного и белого), у которой автоматически подобрано значение порога для выделения двух однородных областей.

#### Введем функцию для поворота изображений, чтобы потом унифицировать направления градиента у особенностей.

```python
def rotateImage(square_image, angle):
    image_size = square_image.shape[0]
    half_width = half_height = half_size = image_size * 0.5
    center = (half_size, half_size)
    rotation_matrix = np.vstack([cv2.getRotationMatrix2D((half_size, half_size), angle, 1.0), [0, 0, 1]])
    rotation_matrix_notranslate = np.matrix(rotation_matrix[0:2, 0:2])
    top_left = (np.array([-half_width, half_height]) * rotation_matrix_notranslate).A[0]
    top_right = (np.array([half_width, half_height]) * rotation_matrix_notranslate).A[0]
    bottom_left = (np.array([-half_width, -half_height]) * rotation_matrix_notranslate).A[0]
    bottom_right = (np.array([half_width, -half_height]) * rotation_matrix_notranslate).A[0]
    
    x_coordinates = [point[0] for point in [top_left, top_right, bottom_left, bottom_right]]
    x_positive = [x for x in x_coordinates if x > 0]
    x_negative = [x for x in x_coordinates if x < 0]
    
    y_coordinates = [point[1] for point in [top_left, top_right, bottom_left, bottom_right]]
    y_positive = [y for y in y_coordinates if y > 0]
    y_negative = [y for y in y_coordinates if y < 0]
    
    right_bound = max(x_positive)
    left_bound = min(x_negative)
    top_bound = max(y_positive)
    bottom_bound = min(y_negative)

    new_width = int(abs(right_bound - left_bound))
    new_height= int(abs(top_bound - bottom_bound))
    new_image_size = (new_width, new_height)
    
    new_center_x = new_width * 0.5
    new_center_y = new_height * 0.5
    
    delta_x = int(new_center_x - half_width)
    delta_y = int(new_center_y - half_height)
    
    translation_matrix = np.matrix([[1, 0, delta_x], [0, 1, delta_y], [0, 0, 1]])
    affine_matrix = (np.matrix(translation_matrix) * np.matrix(rotation_matrix))[0:2, :]
    rotated = cv2.warpAffine(square_image, affine_matrix, new_image_size, flags=cv2.INTER_LINEAR)

    return rotated
```

#### Для начала для наглядности посмотрим, какие особенности выделяются на одном изображении, кластеризуем их. Как дополнительную возможность добавлю отбор по характерному размеру особенности, рассматривать особенности с размером меньше 1-2 пикселей мне кажется странно.

```python
image = cv2.imread("vdata/image_0001.jpg")        
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

sift = cv2.SIFT(150)
keypoints = sift.detect(gray_image,None)
print "All keypoints number -", len(keypoints)
big_keypoints = []
for point in keypoints:
    if point.size > 2:
        big_keypoints += [point]
key_image = cv2.drawKeypoints(gray_image,big_keypoints, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
keypoints, descriptors = sift.compute(gray_image, big_keypoints)
print "Descriptors shape -",descriptors.shape 

print "Визуализация особенностей:"
skimage.io.imshow(key_image)
skimage.io.show()

clustering = KMeans(n_clusters=10)
clusterized_descriptors = clustering.fit_predict(descriptors)
print clusterized_descriptors
```
