## Лабораторная работа №10. Подсчет объектов на видео. Шалашугиной Елены, заочное - CS

```python
import cv2
import cv2.cv as cv
from skimage import io
import numpy
import math
```

Считываем изображения из видеопотока.

```python
frames = []
colored_frames = []
capture = cv2.VideoCapture("visiontraffic.avi")

while True:
    ret, frame = capture.read()
    if ret == 0:
        break
    colored_frames += [frame.copy()]
    grey_frame = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
    frames += [grey_frame]
print len(frames), frames[0].shape
```

### 1. Подсчет фона и отображение видеопотока после вычитания фона.

```python
fgbg = cv2.BackgroundSubtractorMOG2()
fgmask = None
fourcc = cv.CV_FOURCC('M','J','P','G')
writer = cv2.VideoWriter('cars_only.avi', fourcc, 24, (640, 360), False)
masks = []

for frame in frames:
    fgmask = fgbg.apply(frame)
    res = cv2.bitwise_and(frame, frame, mask = fgmask)
    writer.write(res)
    
    masks += [fgmask]
```

Получаем на видео изображения всех машин, их теней и шума в некоторых областях. Видео прикреплено, название 'cars_only.avi'.

### 2. Добавление к исходному видео ограничивающих прямоугольников.

```python
fourcc = cv.CV_FOURCC('M','J','P','G')
writer = cv2.VideoWriter('contours.avi', fourcc, 24, (640, 360), False)
cars = []

for i in xrange(len(masks)):
    mask = masks[i].copy()
    image = frames[i].copy()
    (contours, _) = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    
    contours = filter(lambda x: cv2.contourArea(x) > 1000, contours)
    
    new_cars = []
    
    for contour in contours:
        (x,y,w,h) = cv2.boundingRect(contour)
        cv2.rectangle(image, (x, y), (x + w, y + h), 255)
        new_cars += [(x,y,w,h)]
        
    cars += [new_cars]
    writer.write(image)
    
cars = cars[1:]
```

Из-за шумов изначально находится большое количество контуров. Если убрать совсем маленькие(шумовые), и посмотреть на оставшиеся площади, ограниченные контурами,то можно найти такую площадь, что большие контуры будут ограничивать всю машину(с тенью). Получили видео 'contours.avi' с ограничивающими машины прямоугольниками.

### 3. и 4. Подсчет количества проезжающих машин, всех и только едущих прямо.

```python
current_cars = []
total_cars = 0
forward_cars = 0

for frame_cars in cars:
    previous_cars = current_cars
    current_cars = []
    
    for frame_car in frame_cars:
        (frame_x,frame_y,frame_w,frame_h) = frame_car
        frame_x += frame_w / 2.
        frame_y += frame_h / 2.
        
        found = False
        
        for car in previous_cars:
            (x,y,w,h) = car
            x += w / 2.
            y += h / 2.
            
            distance = math.sqrt(pow(x - frame_x, 2) + pow(y - frame_y, 2))
            
            if distance < 50.:
                found = True
                
        if not found:
            total_cars += 1
            
            if frame_x > 400:
                forward_cars += 1
        
        current_cars += [frame_car]
        
print "Общее количество машин -", total_cars
print "Количество проехавших прямо машин -", forward_cars
            
```

Будем записывать позиции ограничивающих машины прямоугольников на данном кадре и предыдущем. При движении машина изменяет свое положение от кадра к кадру не слишком сильно, не больше какого-то расстояния. Можно находить расстояния между всеми прямоугольниками на текущем и прошлом кадре, если посмотреть на эти числа, можно оценить максимальное расстояние (здесь 50), соответствующее перемещению одной и той же машины. Если не смогли найти соответствие положению машины с предыдущим кадром, значит она только что появилась на экране. Каждая машина въезжает в кадр один раз, просто посчитаем количество таких появлений, получим число всех проехавших машин. При этом заодним определили, в какой координате x машина появилась в кадре. Все машины, поворачивающие направо, делают это из отделенного сплошной правого ряда, то есть заезжают в кадр левее некоторой координаты. Отбросим все машины левее этой точки, получим количество проехавших прямо машин. Результаты совпали с подсчетом вручную.
