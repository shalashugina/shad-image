### 1) Вычисление матрицы диспаритета.

Считываем и показываем изображения.

```python
first_image = cv2.cvtColor(cv2.imread("tsukuba1.png"), cv2.COLOR_BGR2RGB)
second_image = cv2.cvtColor(cv2.imread("tsukuba2.png"), cv2.COLOR_BGR2RGB)
pyplot.figure(figsize=(18, 6))
pyplot.subplot(1, 3, 1)
pyplot.imshow(first_image)
pyplot.subplot(1, 3, 2)
pyplot.imshow(second_image)
print first_image.shape, type(first_image[100][100][0]
```

Будем использовать SAD.

```python
def find_sad(image_first, image_second, line, row_1, row_2, window_half):
    sad = 0
    for il in xrange(-window_half, window_half + 1):
        for jr in xrange(-window_half, window_half + 1):
            for color in xrange(3):
                sad += abs(int(image_first[line + il][row_1 + jr][color]) - int(image_second[line + il][row_2 + jr][color]))
    return sad    
```

Посчитаем оптимальный диспаритет. Если для какого-то диспаритета окрестность пикселя не влезает в рамки изображения, то он убирается из рассмотрения на оптимальность.

```python
def find_disparity(image_first, image_second, line, row, window_half):
    best_d = 0
    min_sad = 1000000
    for d in xrange(31):
        current_sad = 1000000
        if (0 <= line - window_half) and (line + window_half < image_first.shape[0]) and (0 <= row - window_half - d)\
                                     and (row + window_half < image_first.shape[1]):
            current_sad = find_sad(image_first, image_second, line, row, row - d, window_half);
        if current_sad < min_sad:
            min_sad = current_sad
            best_d = d
            
    return best_d 
```

Получим карту, найдя оптимальный диспаритет для каждой точки изображения.

```python
def make_disparity_map(image_first, image_second, window_half):    
    result_map = np.zeros((image_first.shape[0], image_first.shape[1]))
    for line in xrange(image_first.shape[0]):
        for row in xrange(image_first.shape[1]):
            result_map[line, row] = find_disparity(image_first, image_second, line, row, window_half)
    return result_map
```

Считаются результаты очень долго и, естественно, квадратично от размера стороны окна.

```python
map_results = []

for half_size in xrange(6):
    print "Длина стороны окна -", (half_size * 2 + 1)
    current_disparity_map = make_disparity_map(first_image, second_image, half_size)
    map_results += [current_disparity_map]
    pyplot.imshow(current_disparity_map, cmap = 'gray')
    pyplot.show()
```

Для длины стороны окна 7 еще хорошо сохранены детали, а шумов уже заметно меньше, чем со стороной 3, считаю его оптимальным размером окна.

### 2) Вычисление матрицы диспаритета внутренними методами

Параметр ndisparities отвечает за проверяемый диапазон диспаритета. Параметр SADWindowSize задает сторону сравниваемых блоков.

```python
first_grey_image = cv2.cvtColor(first_image, cv2.COLOR_RGB2GRAY)
second_grey_image = cv2.cvtColor(second_image, cv2.COLOR_RGB2GRAY)
```

```python
def show_disparity_map(first_grey_image, second_grey_image, n_dis, window_size):
    disp = cv2.StereoBM(preset = cv2.STEREO_BM_BASIC_PRESET, ndisparities = n_dis, SADWindowSize = window_size)\
                    .compute(first_grey_image, second_grey_image)
    pyplot.imshow(disp, cmap = 'gray')
    pyplot.show()
```

Посмотрим, как влияет ndisparities и SADWindowSize на изображение.

```python
parameters = [(16, 5), (32, 5), (64, 5), (32, 7), (32, 9),(32, 11), (32, 13),(32, 19),]

for n, size in parameters:
    print n, size
    show_disparity_map(first_grey_image, second_grey_image, n, size)
```

Большое ndisparities вызывает обрезание заметного края изображения слева. В остальном влияние ndisparities незаметно, меняются значения в карте(контраст изображения), изображения с большим ndisparities темнее. Скорее всего в этой функции compute введена нормировка(деление) на максимальное значение ndisparities, тогда при тех же сдвигах точки мы будем получать меньшее значение на карте и более темный цвет на изображении. 

Мне кажется здесь хорошим значением параметр SADWindowSize = 11, шумов не очень много и однородные области показаны довольно хорошо. Сравним это изображение и полученное ранее для стороны 7.

```python
print "Second part result:"
show_disparity_map(first_grey_image, second_grey_image, 32, 11)
print "First part result:"
pyplot.imshow(map_results[3], cmap = 'gray')
pyplot.show()
```
