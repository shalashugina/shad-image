Импортируем необходимые библиотеки

```python
import skimage.io
from matplotlib import pyplot
```

#### 1. Считываем изображение, создаем окно для вывода и выводим изображение.

```python
image = skimage.io.imread("duckling.png")
skimage.io.imshow(image)
skimage.io.show()
```

Ниже приведен скриншот окна с изображением, полученный до добавления % mathplotlib inline.

```python
%matplotlib inline
screenshot = skimage.io.imread("screenshot.png")
skimage.io.imshow(screenshot)
skimage.io.show()
```

####  2. Выводим размер изображения. Создаем изображения из верхней и левой половинок, сохраняем их.

```python
print "Размер изображения - ", image.shape

top = image[0:image.shape[0]/2,:,:]
pyplot.imsave("top.png", top)
skimage.io.imshow(top)
skimage.io.show()

left = image[:,0:image.shape[1]/2,:]
pyplot.imsave("left.png", left)
skimage.io.imshow(left)
skimage.io.show()
```

####  3. Выведем изображения для каждого из RGB каналов и без каждого из каналов.

```python
colors = ((0, 'red'), (1, 'green'), (2, 'blue'))

for color, name in colors:
    color_image = image.copy()
    for other_color in (0, 1, 2) : 
        if other_color != color :
            color_image[:,:,other_color] = 0    
    skimage.io.imshow(color_image)
    print name
    skimage.io.show()
    
for color, missing_name in colors:
    color_image = image.copy()
    color_image[:,:,color] = 0
    skimage.io.imshow(color_image)
    print "without", missing_name
    skimage.io.show()
```

#### 4. Вычислим усредненное по каналам изображение. Вставим в отчет его и сконвертированное в черно-белое.

```python
copy_image = image.copy()
averaged_image = copy_image[:,:,3]

for color, name in colors:
    averaged_image += image[:,:,color] / 3

skimage.io.imshow(averaged_image)
print "Averaged image"
skimage.io.show()

grey_image = skimage.color.rgb2gray(image)
skimage.io.imshow(grey_image)
print "Converted image"
skimage.io.show()
```

Сконвертированное изображение выглядит более ярко, контрастно. Можно заметить, что второе изображение схоже с Value каналом HSV. Этот канал выбирается как максимум значений яркости цветов из RGB. А при усреднении яркость получается ниже, если какого-то цвета в данном оттенке мало, что особенно заметно для основных цветов. Возможно, что человек воспринимает скорее яркость какого-то отдельного цвета, чем яркость в среднем, поэтому важнее максимально яркий цвет, как в HSV, а не усредненные каналы RGB.

#### 5. Преобразуем изображение из RGB в HSV. Выводим изображения отдельных каналов.

```python
hsv_image = skimage.color.rgb2hsv(image[:,:,0:3])
for channel in (0, 1, 2):
    channel_image = hsv_image[:,:,channel]
    skimage.io.imshow(channel_image)
    skimage.io.show()
```

#### 6. Получаем негатив черно-белого изображения.

```python
print skimage.dtype_limits(grey_image)
import numpy
ones = numpy.ones_like(grey_image)
invert_image = ones - grey_image
skimage.io.imshow(invert_image)
skimage.io.show()
```

#### 7. Зеркально отобразим изображение относительно вертикальной оси.

```python
mirrored_image = image.copy()
mirrored_image = image[:,::-1,:]
skimage.io.imshow(mirrored_image)
skimage.io.show()
```
