```python
import skimage.io
from sklearn.decomposition import PCA
from sklearn.neighbors import NearestNeighbors
import skimage.feature
import scipy.io
%matplotlib inline
from matplotlib import pyplot
import numpy
import math
```

## Распознавание лиц.

Считываем данные.

```python
f_matrix = scipy.io.loadmat("faces_data.mat")
print f_matrix.keys()

train = f_matrix['train_faces'].transpose()
train_labels = f_matrix['train_labels']
test = f_matrix['test_faces'].transpose()
test_labels = f_matrix['test_labels']

print f_matrix['__header__'], f_matrix['__globals__'], f_matrix['__version__']
print train.shape, train_labels.shape, test.shape, test_labels.shape
```

### 1. Распознавание на основе метода ближайшего соседа.

```python
NN = NearestNeighbors(n_neighbors=1).fit(train)
distances, test_answers = NN.kneighbors(test)

count = 0
for test_number in xrange(len(test_labels)):
    if train_labels[test_answers[test_number]] == test_labels[test_number]:
        count += 1
    else:
        pass
        #print train_labels[test_answers[test_number]], test_labels[test_number]

print "Количество верных ответов -", count        
print "Процент верных ответов -", (count*100.0)/len(test_labels)
```

Можно было бы визуализовать часть пар сопоставленных изображений для проверки, но занимает много места в отчете. Качество метода около 89 процентов.

### 2. Применим подход Eigenfaces к обучающей базе.

```python
pca = PCA()
pca.fit(train)
average = pca.mean_
skimage.io.imshow((average.reshape(32, 32)).transpose())
skimage.io.show()
```

Визуализовали "усредненное лицо".

### 3. Визуализуем восстановленные фотографии с разным количеством главных компонент.

```python
def visualize_restored_images(a):    
    components_numbers = [1, 2, 5, 10, 100]
    print  '\n', '\n', "Исходное изображение"
    print "Номер", a
    skimage.io.imshow((train[a].reshape(32, 32)).transpose())
    skimage.io.show()
    for number in components_numbers:
        pca = PCA(n_components=number)
        pca.fit(train)
        comp_train = pca.transform(train)
        restored_train = pca.inverse_transform(comp_train)
        print "Восстановленное с использованием", number, "компонент"
        skimage.io.imshow((restored_train[a].reshape(32, 32)).transpose())
        skimage.io.show()
visualize_restored_images(11)
visualize_restored_images(50)
```

Одно из изображений неплохо восстанавливается уже при 10 компонентах, другое хуже из-за редкого ракурса (сначала попытка восстановить ракурс, а не черты лица). Со ста компонентами отличий от исходного мало. С одной-двумя компонентами, наоборот, получается усредненное по базе лицо, независимое от исходного.

### 4. Распознавание с использованием K главных компонент для описания изображения.

```python
K_number = [1, 2, 3, 4, 5, 7, 9, 11, 13, 15, 17, 20, 25, 30, 35, 40, 50, 60, 80, 100]
K_counts = []
max_count = 0
for K in K_number:
    pca = PCA(K)
    pca.fit(train)
    comp_train = pca.transform(train)
    comp_test = pca.transform(test)
    NN = NearestNeighbors(n_neighbors=1).fit(comp_train)
    distances, test_answers = NN.kneighbors(comp_test)
    count = 0
    for test_number in xrange(len(test_labels)):
        if train_labels[test_answers[test_number]] == test_labels[test_number]:
            count += 1
    if count > max_count:
        max_count = count
    print K, count
    K_counts += [count]
    restored_train = pca.inverse_transform(comp_train)
pyplot.scatter(K_number, K_counts)
print "Наилучший процент распознавания -", 100.0*max_count/len(test_labels)
```

При K=17 достигается то же количество правильных ответов, что и на несжатых изображениях. Такой поиск может быть лучше, так как иногда отбрасываем несущественную, шумовую информацию. В данном случае процент распознаваний получается чуть лучше, чем наивным методом, 90%.

### 5. Распознавание лиц с помощью гистограмм LBP признаков.

```python
def compute_LBP_hist(image_data, n_points, radius):
    image = image_data.reshape(32, 32).transpose()
    LBP_map = skimage.feature.local_binary_pattern(image, n_points, radius)
    n_bins = LBP_map.max() + 1
    hist, _ = numpy.histogram(LBP_map, bins = n_bins, range = (0, n_bins), normed=True)
    hist_2 = hist.copy()
    for i in xrange(len(hist)):
        hist_2[i] = math.sqrt(hist[i])
    return hist_2

def hist_for_base(base, n_points, radius):
    hist_base = numpy.ndarray((0, pow(2, n_points)))
    for im in base:
        hist_im = compute_LBP_hist(im, n_points, radius)
        hist_base = numpy.vstack((hist_base, hist_im))
    return hist_base

n_points = 8
radius = 4

hist_train = hist_for_base(train, n_points, radius)
hist_test = hist_for_base(test, n_points, radius)

NN = NearestNeighbors(n_neighbors=1).fit(hist_train)
distances, test_answers = NN.kneighbors(hist_test)

count = 0
for test_number in xrange(len(test_labels)):
    if train_labels[test_answers[test_number]] == test_labels[test_number]:
        count += 1
        
print "Test length -", len(hist_test), ";    Right test answers -", count
print "Procent of right answers - ", (count *100.0)/len(test_labels)
```

Можно подбирать параметры радиуса и количесва точек на окружности для улучшения результата. Поэлементное извлечение корня для гистограмм улучшает результат (например с 114 до 117 обнаружений из 120 для n_points = 8, radius = 4). С фиксированным радиусом 4 один из лучших результатов при 8 точках, переберем другие радиусы с этим количеством точек. Радиус 3 дает только 111 правильных ответов, 2 - 109, 5 - 113, 6 - 111. Увеличение количества точек до 12 улучшает результат радиусов 6, 5 и 3 до 114, 117 и 116. 

В итоге считаю, что результат распознавания 97.5% процентов достаточно хорош, заметно лучше 90% метода Eigenfaces.
