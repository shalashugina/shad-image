## Лабораторная работа №10. Шалашугиной Лены - заочное(CS)

```python
import cv2
from matplotlib import pyplot
from skimage import feature
import matplotlib
import numpy
from skimage.measure import ransac
from skimage import transform 
%matplotlib inline
```

```python
first_image = cv2.cvtColor(cv2.imread('demo1.png'), cv2.COLOR_BGR2GRAY)
second_image = cv2.cvtColor(cv2.imread('demo2.png'), cv2.COLOR_BGR2GRAY)
print first_image.shape, second_image.shape
```

### 1) Вычисление дескрипторов.

```python
def find_keypoints_and_descriptors(image, image_name):
    sift = cv2.SIFT(1000)
    keypoints, descriptors = sift.detectAndCompute(image, None)
    key_image = cv2.drawKeypoints(image, keypoints, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
    
    sk_keypoints = numpy.zeros((0, 2)) 
    for i in xrange(len(keypoints)):
        sk_keypoints = numpy.vstack((sk_keypoints, (keypoints[i].pt) ))
           
    print "Визуализация особенностей -", image_name 
    matplotlib.rc("savefig", dpi=100)
    pyplot.imshow(key_image)
    pyplot.show()
    return sk_keypoints, descriptors
```

```python
first_keypoints, first_descriptors = find_keypoints_and_descriptors(first_image, "Первое изображение")
second_keypoints, second_descriptors = find_keypoints_and_descriptors(second_image, "Второе изображение")
```

Увеличение количества находимых особенностей с 1000 до 2000 повышает количество найденных соответствий со 184 до 259 без геометрической валидации, и с около 90 до около 140 с ней. Качество панорамы от этого не зависит.

### 2) Сопоставление дескрипторов.

```python
matches = feature.match_descriptors(first_descriptors, second_descriptors, cross_check=True)
print "Число найденных соответствий:", matches.shape[0]

# Отчего-то если передавать координаты (x,y) особых точек в изначальном порядке, получается неверно тут, иначе в вычислении model у ransac
first_keypoints_inverted = numpy.array([(first_keypoints[i][1], first_keypoints[i][0]) for i in xrange(first_keypoints.shape[0])])
second_keypoints_inverted = numpy.array([(second_keypoints[i][1], second_keypoints[i][0]) for i in xrange(second_keypoints.shape[0])])

matplotlib.rc("savefig", dpi=180)
fig = pyplot.figure()
ax = fig.add_subplot(111)
feature.plot_matches(ax, first_image, second_image, first_keypoints_inverted, second_keypoints_inverted, matches, keypoints_color='c', only_matches=True)
ax.axis('off')
pyplot.show()
```

Видны линии сопоставлений, направленные отлично от основной части соответствующих точек, соединяющие заведомо различные точки, идущие в несовпадающие части фотографий. Но уже отсеяно большое количество особых точек, для которых не нашлось взаимно ближайшей точки.

### 3) Геометрическая валидация.

```python
source_coords = numpy.zeros((0, 2))
dist_coords = numpy.zeros((0, 2))
for i in xrange(matches.shape[0]):
    source_coords = numpy.vstack((source_coords, first_keypoints[matches[i][0]]))    
    dist_coords = numpy.vstack((dist_coords, second_keypoints[matches[i][1]]))

transform_model, inliers = ransac((source_coords, dist_coords), model_class=transform.ProjectiveTransform, min_samples=4, residual_threshold=3)
print "Матрица преобразования:\n", transform_model.params

new_matches = numpy.array([matches[i] for i in xrange(len(inliers)) if inliers[i] == True])
print "Оставшееся количество соответствий -", new_matches.shape[0]

matplotlib.rc("savefig", dpi=180)
fig = pyplot.figure()
ax = fig.add_subplot(111)
feature.plot_matches(ax, first_image, second_image, first_keypoints_inverted, second_keypoints_inverted, new_matches, keypoints_color='c', only_matches=True)
ax.axis('off')
pyplot.show()
```

Теперь остались соответствия, прошедшие геометрическую валидацию. Линии, показывающие их, имеют близкие длину и направление. Сопоставляемые точки находятся в областях объектов, являющихся общими у изображений.

### 4) Визуализация панорамы.

```python
#Проконтролируем, во что перешли угловые точки
corners = numpy.array([(0, 0), (0, second_image.shape[0]), (second_image.shape[1], 0), (second_image.shape[1], second_image.shape[0])])
print  "Corners:\n", corners
new_corners = transform_model.inverse(corners)
print "\n New corners positions:\n", new_corners, '\n'

#Определим, где проходят новые границы второго изображения
x_min = min(new_corners[:, 0])
print "X_min =", x_min
x_max = max(new_corners[:, 0])
print "X_max =", x_max
y_min = min(new_corners[:, 1])
print "Y_min =", y_min
y_max = max(new_corners[:, 1])
print "Y_max =", y_max, '\n'

#Находим новые размеры изображения, сравниваем с размерами первого, находим необходимый сдвиг
y_size = y_max - min(y_min, 0)
x_size = x_max - min(x_min, 0)
print "  X_size =", x_size, first_image.shape[1]
print "  Y_size =", y_size, first_image.shape[0]
offset = transform.SimilarityTransform(translation=[int(min(x_min, 0)), int(min(y_min, 0))])

#Применяем преобразования к изображениям
new_second_image = transform.warp(second_image, offset + transform_model, output_shape=(int(y_size), int(x_size)))
new_first_image = transform.warp(first_image, offset, output_shape=(int(y_size), int(x_size)))

#Находим область пересечения, совмещаем изображения с ее учетом
mask = (new_first_image != 0) & (new_second_image != 0)
result_image = new_first_image + new_second_image
result_image[mask] /= 2

pyplot.imshow(result_image, cmap='gray')
pyplot.show()
```
