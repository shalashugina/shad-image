### Задание 1. Различение изображений рукописных цифр.

#### 1. Загрузка данных.

```python
digits = datasets.load_digits()
print digits.data.shape, digits.target.shape, digits.target_names
digits_data = digits.data
digits_labels = digits.target
print '\n', "Пример изображения цифры"
pylab.matshow(digits.images[0]) 
pylab.show() 
```

Перемешивание данных в случайном порядке.

```python
indexes = range(len(digits_data))
random.shuffle(indexes)
d_data = [ digits_data[x] for x in indexes ]
d_labels = [ digits_labels[x] for x in indexes ]
d_images = [digits.images[x] for x in indexes]
```

#### 2. Кластеризация изображений с помощью KMeans. Предполагаю, что под запуском KMeans при различных начальных приближениях имеется ввиду несколько циклов обучения и предсказания для параметров n_init=1 и init='random' (разбиение на кластеры одним запуском при случайном начальном приближении).

```python
def clasterize(data, labels, verbose):
    kmeans_clasterisation = KMeans(n_clusters=10, n_init=1, n_jobs=-1, init='random')
    new_labels = kmeans_clasterisation.fit_predict(data)
    AMI_score = metrics.adjusted_mutual_info_score(labels, new_labels)
    ARI_score = metrics.adjusted_rand_score(labels, new_labels)
    if verbose:
        print "Adjusted Mutual Information -", AMI_score
        print "Adjusted Rand index -", ARI_score
    return (AMI_score, ARI_score, kmeans_clasterisation.cluster_centers_, new_labels)
```

Выбрано n_clusters=10, потому что это дает необходимое количество классов (хотя при 11 оценки качества лучше).

#### 3. Проведем кластеризации и выведем оценки их качества. Заодним запомним центроиды для лучшей из кластеризаций.

```python
best_centroids = numpy.ndarray
best_new_labels = numpy.ndarray
max_AMI = 0

for i in xrange(1, 6):
    print '\n', "Запуск", i
    AMI_score, ARI_score, centroids, new_labels = clasterize(d_data, d_labels, True)
    if (AMI_score >= max_AMI):
        max_AMI = AMI_score
        best_centroids = centroids   
        best_new_labels = new_labels
```

Виден заметный разброс результатов в зависимости от начальных условий. 

Проверим, одинакова ли лучшая классификация для разных мер и насколько хороший результат можно получить при большом количестве запусков.

```python
best_AMI_centroids = numpy.ndarray
best_ARI_centroids = numpy.ndarray
best_new_labels = numpy.ndarray
max_AMI = 0
max_ARI = 0
min_AMI = 1
min_ARI = 1

for i in xrange(1,500):
    AMI_score, ARI_score, centroids, new_labels = clasterize(d_data, d_labels, False)
    
    if (AMI_score >= max_AMI):
        max_AMI = AMI_score
        best_AMI_centroids = centroids
        best_new_labels = new_labels
    if (ARI_score >= max_ARI):
        max_ARI = ARI_score
        best_ARI_centroids = centroids 
        best_new_labels = new_labels
        
    if (AMI_score < min_AMI):
        min_AMI = AMI_score
    if (ARI_score < min_ARI):
        min_ARI = ARI_score
        
print "min_AMI", min_AMI, "max_AMI", max_AMI
print "min_ARI", min_ARI, "max_ARI", max_ARI

if best_AMI_centroids.all() == best_ARI_centroids.all():
    print '\n', "One best result"
```

Лучшая кластеризация одна и та же для обеих метрик(проверено более чем на 500 случайных запусках), поэтому можно было заводить только одну переменную для наилучшего результата. Лучший результат на 8% лучше среднего по AMI, на 14% по ARI.

#### 4. Визуализация изображений, соответствующих центроидам лучшей кластеризации.

#### 5. Примеры ошибочно отнесенных объектов.

Посчитаем средний класс объектов в каждом кластере.

```python
counter = numpy.zeros((10, 10))
for i in xrange(len(best_new_labels)):
    cluster_number = best_new_labels[i]
    label = d_labels[i]
    counter[cluster_number][label] += 1
    
average_labels = numpy.zeros(10)
average_count_labels = numpy.zeros(10)
for i in xrange(10):
    for j in xrange(10):
        if average_count_labels[i] < counter[i][j]:
            average_count_labels[i] = counter[i][j]
            average_labels[i] = j
print average_labels, "- Метки большей части объектов кластера"
print average_count_labels, "- Количество объектов с правильной меткой в кластере"
```

Есть классы, соответствующие каждому из чисел. Для каждого из чисел в выборке было порядка 180 примеров, довольно большая часть отнесена правильно.

Теперь собственно приведем примеры ошибочных классификаций для некоторых кластеров.

```python
def find_bad_object(cluster_number):
    average_label = average_labels[cluster_number]
    for i in xrange(len(best_new_labels)):
        number = best_new_labels[i]
        label = d_labels[i]
        if number == cluster_number:
            if label != average_label:
                return d_images[i]
            
clusters_n = [0, 1, 3, 4, 7, 8]          
for i in clusters_n:
    print '\n', "Ошибочная классификация в кластер №", i, "с меткой", int(average_labels[i])
    pylab.matshow(find_bad_object(i)) 
    pylab.show()
```

Ошибки довольно понятны и даже на глаз некоторые цифры разобрать сложновато.

### Задание 2. Уменьшение палитры цветов изображения с минимальными потерями визуального качества.

#### 1. Считываем изображение.

```python
Image = io.imread('http://www.birdsgallery.net/gallery/parrots/parrots_4.jpg')
```

Смотрим на различные каналы изображения.

```python
pyplot.figure(figsize=(15, 5))
for i in xrange(3):
    pyplot.subplot(1, 3, i)
    io.imshow(Image[:,:,i])
```

#### 2. Научимся выделять кластеры и заменять все цвета из кластера на какой-то средний (центроид или медиана).

Соберем все встретившиеся точки нужного пространства

```python
pixels_rgb = []
pixels_coord = []
for i in xrange(Image.shape[0]):
    for j in xrange(Image.shape[1]):
        pixel = Image[i][j]
        pixels_rgb += [pixel]
        pixels_coord += [(i, j)]
print len(pixels_rgb)
```

Введем функцию для кластеризации, запустим ее.

```python
def KMeans_clusterize_rgb(pixels,n_clusters):
    kmeans_clasterisation = KMeans(n_clusters=n_clusters, n_init=10, n_jobs=-1)
    cluster_number = kmeans_clasterisation.fit_predict(pixels)
    return (kmeans_clasterisation.cluster_centers_, cluster_number)

rgb_centroids, rgb_clusters = KMeans_clasterize_rgb(pixels_rgb, 10)
print rgb_centroids
```

Введем функцию для смены цвета пикселов на изображении, проверим ее.

```python
def clacterize_image(image, pixels_coord, clusters, clusters_centres):
    cl_image = image.copy()
    for n_cl in xrange(len(clusters)):
        i, j = pixels_coord[n_cl]
        cluster_number = clusters[n_cl]
        cl_image[i][j] = clusters_centres[cluster_number]
    return cl_image

new_image = clacterize_image(Image, pixels_coord, rgb_clusters, rgb_centroids)

pyplot.figure(figsize=(15, 5))
for i in xrange(2):
    pyplot.subplot(1, 2, i)
    if i == 0:
        pyplot.title("New")
        io.imshow(new_image)
    else:
        pyplot.title("Old")
        io.imshow(Image)
```

#### 3. Выполним кластеризацию с различным числом кластеров.

```python
numbers = [2, 3, 10, 20]
images = []

for number in numbers:
    rgb_centroids, rgb_clusters = KMeans_clasterize_rgb(pixels_rgb, number)
    new_image = clacterize_image(Image, pixels_coord, rgb_clusters, rgb_centroids)
    images += [new_image]
```

Визуализируем результат

```python
pyplot.figure(figsize=(20, 4))
for i in xrange(1, 6):
    pyplot.subplot(1, 5, i)
    if i == 5:
        pyplot.title("Initial image")
        io.imshow(Image)
    else:
        pyplot.title("Image with {} color clusters".format(numbers[i-1]))
        io.imshow(images[i-1])
```

#### 5. Оценка потери качества

```python
def find_quality_loss(initial_image, new_image):
    MSE = ((initial_image - new_image) ** 2).mean(axis= None)
    MAX_I = 255.
    PSNR = 10*math.log10(MAX_I**2 / MSE)
    return PSNR

for i in xrange(4):
    image = images[i]
    print "PSNR для", numbers[i], "кластеров -", find_quality_loss(image, Image)
```

#### 2. Выделение сегментов с помощью кластеризации.

Составим вектор признаков, соответствующий всем пикселям изображения.

```python
def find_features(lambd, image):
    features = []    
    for i in xrange(image.shape[0]):
        for j in xrange(image.shape[1]):
            pixel = image[i][j]
            features += [[lambd*i, lambd*j, pixel[0], pixel[1], pixel[2]]]
    return features   
print len(find_features(0.5, im))
```

```python
def KMeans_clusterize(features,n_clusters):
    kmeans_clasterisation = KMeans(n_clusters=n_clusters, n_init=10, n_jobs=-1)
    cluster_number = kmeans_clasterisation.fit_predict(features)
    return (kmeans_clasterisation.cluster_centers_, cluster_number)

KM_features = find_features(1, im)
centroids, KM_clusters = KMeans_clusterize(KM_features, 10)
```
