#### 2. Создадим отдельную функцию houghlines.

```python
import math

def houghlines(image, distance_step, degree_step, accuracy):
    R_steps = (int)((math.sqrt(image.shape[0]**2 + image.shape[1]**2))/distance_step)
    angle_steps =  360 / degree_step
    matrix = scipy.zeros((R_steps, angle_steps), scipy.uint16)
    print "R_steps -", R_steps,"; angle_steps -", angle_steps
    
    for x in xrange(image.shape[0]):
        for y in xrange(image.shape[1]):
            if image[x][y] > 0.5:
                for R in range(R_steps):
                    distance = (R + 0.5) * distance_step
                    for angle in range(angle_steps):
                        radian_angle = 2 * 3.1415 * (angle + 0.5)/ angle_steps
                        if abs(y * math.sin(radian_angle) + x * math.cos(radian_angle) - distance) < accuracy:
                            matrix[R][angle] += 1   #считаем точку, попавшую на данную прямую              
   
    return matrix
```

#### 2. Запишем параметрическое уравнение окружности. (x0, y0) - центр окружности, R - ее радиус. Уравнение выглядит как [(x - x0)^2 + (y - y0)^2 = R^2]

#### 3.Напишем функцию houghcircles. Такое количество вложенных циклов даже выглядит подозрительно.

```python
def houghcircles(image, distance_step, radius_step, accuracy):
    R_steps = (int)((math.sqrt(image.shape[0]**2 + image.shape[1]**2)) / radius_step)
    distance_steps_number =  (int)((math.sqrt(image.shape[0]**2 + image.shape[1]**2)) / distance_step)
    matrix = scipy.zeros((R_steps, distance_steps_number, distance_steps_number), scipy.uint16)
    
    for x in xrange(image.shape[0]):
        for y in xrange(image.shape[1]):
            if image[x][y] > 0.5:
                for R in range(R_steps):
                    R_distance = (R + 0.5)* radius_step
                    for x0 in range(distance_steps_number):
                        x_distance = (x0 + 0.5) * distance_step
                        for y0 in range(distance_steps_number):
                            y_distance = (y0 + 0.5) * distance_step
                            if abs((y - y_distance)**2 + (x - x_distance)**2 - R_distance**2) < accuracy:
                                matrix[R][x0][y0] += 1               
   
    return matrix
```
