### Первая часть. Изображение Сатурна

#### 1. Преобразование Фурье изображения

```python
saturn_image = skimage.io.imread("saturn.jpg")
print "Изначальное изображение"
skimage.io.imshow(saturn_image)
skimage.io.show()

fourier_image = scipy.fftpack.fft2(saturn_image)
fourier_image = scipy.fftpack.fftshift(fourier_image)

amplitude_image = scipy.log(1. + scipy.absolute(fourier_image))
phase_image = scipy.angle(fourier_image)

print "Амплитудный спектр"
skimage.io.imshow(amplitude_image)
skimage.io.show()

print "Фазовый спектр"
skimage.io.imshow(phase_image)
skimage.io.show()
```

Хорошо заметна по Фурье-преобразованию симметрия относительно оси Сатурна. Сама размытая полоса на амплитудном спектре по всей видимости соответствует кольцам Сатурна и кольцам на его поверхности. Горизонтальная полоса спектра относится к вертикальным полосам, являющимся артефактами сжатия или съемки.

#### 2. Зануление низких или высоких частот.

```python
high_frequency_fft = fourier_image.copy()
high_frequency_fft[124:132,124:132] = 0

high_frequency_image = scipy.real(scipy.fftpack.ifft2(high_frequency_fft))

print "Изображение с удаленными нижними частотами"
skimage.io.imshow(high_frequency_image)
skimage.io.show()
```

```python
for x in xrange(1, 6):
    low_frequency_fft = numpy.zeros_like(fourier_image)
    low_frequency_fft[128-x:128+x,128-x:128+x] = fourier_image[128-x:128+x,128-x:128+x]
    
    low_frequency_image = scipy.real(scipy.fftpack.ifft2(low_frequency_fft))
    
    print "Количество оставленных на изображении низких частот - ", x
    skimage.io.imshow(low_frequency_image)
    skimage.io.show()
```

### Вторая часть. Поиск клада.

#### 1. Бинаризуем изображение с сохранением объектов и стрелок. Здесь белыми становились все изображения с более чем 5% яркости.

```python
def binarize_image(image):
    image_gray = skimage.color.rgb2gray(image)
    
    image_gray[image_gray > 0.05] = 1
    image_gray[image_gray < 1] = 0
    
    return image_gray


for image_name in ('Klad00.jpg', 'Klad01.jpg', 'Klad02.jpg'):
    image = skimage.io.imread(image_name)
    print "Initial image", image_name
    skimage.io.imshow(image)
    skimage.io.show()
    
    binary_image = binarize_image(image)
    print "Binarized image"
    skimage.io.imshow(binary_image)
    skimage.io.show()
```

#### 2. Выделяем связные компоненты, визуализируем результаты.

```python
for image_name in ('Klad00.jpg', 'Klad01.jpg', 'Klad02.jpg'):
    image = skimage.io.imread(image_name)
    binary_image = binarize_image(image)
    
    image_label = skimage.measure.label(binary_image)
    print "Визуализация выделения связных компонент", image_name
    skimage.io.imshow(skimage.color.label2rgb(image_label))
    skimage.io.show()
```

#### 3. Функция нахождения стрелок на изображении, стрелки отделяются по занимаемой ими площади.

```python
def is_arrow(item):
    return item.filled_area > 1400 and item.filled_area < 1600

#labels = skimage.measure.regionprops(image_label) - вычисляется внутри основного алгоритма 
def get_arrows(labels):
    return [x for x in labels if is_arrow(x)]
```

#### 4. Функция нахождения начальной красной стрелки

```python
def is_start_arrow(image, arrow):
    coords = scipy.int32(arrow.centroid)
    return image[coords[0], coords[1], 1] < 128

def find_start_arrow(image, arrows):
    for arrow in arrows:
        if is_start_arrow(image, arrow):
            return arrow
    return None
```

#### 5. Направление стрелки определяем с помощью точек, присутствующих ближе к острию стрелки на изначальных изображениях. Направление на эти точки из центра считаем направлением стрелки. Функция поиска клада определяет метку объекта, на который указывает стрелка, если это клад, то останавливается, иначе ищет следующий.

```python
def get_arrow_direction(image, arrow):
    if is_start_arrow(image, arrow):
        functor = lambda x, y: image[x,y,1] > 128
    else:
        functor = lambda x, y: image[x,y,0] < 128
        
    x_sum = 0
    y_sum = 0
    count = 0
    for x, y in arrow.coords:
        if functor(x, y):
            x_sum += x
            y_sum += y
            count += 1
            
    x_sum /= float(count)
    y_sum /= float(count)
    
    center = arrow.centroid
    
    direction = (x_sum - center[0], y_sum - center[1])
    length = math.sqrt(direction[0] ** 2 + direction[1] ** 2)
    return (direction[0] / length, direction[1] / length)
```

#### 6. Итоговый код нахождения клада на изображении, клад выделяется кружочком.

```python
def find_klad(arrow, image, items):
    center = arrow.centroid
    direction = get_arrow_direction(image, arrow)
    
    next_item = None
    next_distance = 1e9
    
    for item in items:
        if item == arrow:
            continue
            
        for x, y in item.coords:
            new_direction = (x - center[0], y - center[1])
            
            new_distance = math.sqrt(new_direction[0] ** 2 + new_direction[1] ** 2)
            if new_distance > next_distance:
                continue
            
            scalar_prod = new_direction[0] * direction[0] + new_direction[1] * direction[1]
            if abs(scalar_prod / new_distance - 1) < 1e-3:
                next_item = item
                next_distance = new_distance
                
    assert(next_item != None)
    
    if is_arrow(next_item):
        find_klad(next_item, image, items)
    else:
        coords = skimage.draw.circle_perimeter(int(next_item.centroid[0]), int(next_item.centroid[1]), 50)
        image[coords] = (255, 255, 255)
        print 'Клад:', scipy.int32(next_item.centroid)
        skimage.io.imshow(image)
        skimage.io.show()
```

Запуск поиска клада для всех изображений.

```python
for image_name in ('Klad00.jpg', 'Klad01.jpg', 'Klad02.jpg'): 
    image = skimage.io.imread(image_name)
    binary_image = binarize_image(image)
    image_label = skimage.measure.label(binary_image)
    
    labels = skimage.measure.regionprops(image_label)
    arrows = get_arrows(labels)
    start_arrow = find_start_arrow(image, arrows)                
    assert(start_arrow != None)
    
    find_klad(start_arrow, image, labels)
```
